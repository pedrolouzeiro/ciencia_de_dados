```python
#df
```

# Limpando os dados

```python
df.drop(['device', 'geoNetwork', 'trafficSource', 'totals', 'adwordsClickInfo'], axis=1, inplace=True)
#df
```

```python
coluna_na = [] 
```

```python
for coluna in df.columns:
    print(coluna + ': ' + str(len(df[coluna].unique())))
    if len(df[coluna].unique()) == 1:
        coluna_na.append(coluna)
```

```python
df.drop(coluna_na, axis=1, inplace=True)
df.head()
```

## Procurar variáveis certas para predição que eu quero

```python
quant = ['bounces', 'hits', 'newVisits', 'pageviews', 'transactionRevenue']
for coluna in quant:
    df[coluna] = pd.to_numeric(df[coluna])
```

### Lembrar de levantar hipóteses para certeza que elas influenciam
ex: pessoas com mais bounces compram mais??

Qual a chance dele parar de comprar ou voltar a comprar?

```python
df['transactionRevenue'] = df['transactionRevenue'].fillna(0)
```

```python
df['transactionRevenue'] = df['transactionRevenue'] / 1000000
```

```python
len(set(df.fullVisitorId))
```

```python
df_quant = df.groupby(['fullVisitorId'], as_index=False)[quant].sum()
```

# Separando as bases

```python
y = df_quant['transactionRevenue'].copy()
X = df_quant.drop('transactionRevenue', axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Regressão Linear

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg_predict= reg.predict(X_test)
```

# Resultados

```python
resultados = pd.DataFrame()
resultados['revenue'] = y_test
resultados['predict'] = reg_predict
resultados['erro'] = reg_predict - y_test
resultados
```

```python
resultados.query('revenue > 0')
```

# RMSE

```python
from sklearn.metrics import mean_squared_error
import numpy as np

np.sqrt(mean_squared_error(y_test, reg_predict))
```
